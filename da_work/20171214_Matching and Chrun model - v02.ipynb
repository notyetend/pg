```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import recall_score, roc_curve

# df = pd.read_csv('20171212_T5_data003_Mobile_T5__log_stat_tracking.csv')  # dataset extracted using query attatched
df = pd.read_csv('20171221_T5_data004_Mobile_T5__log_stat_tracking.csv')
y = df.iloc[:, 0].values
X = df.drop(df.columns[0], axis=1).values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)

class_labels = ['stay', 'churn']
```

###### Performance matrics

```python
from sklearn.metrics import confusion_matrix, recall_score, roc_curve, classification_report
import matplotlib.pyplot as plt
import itertools
% matplotlib inline


class PerfMatrics():
    def __init__(self, y_actual, X, model, class_labels=None):
        self.y_actual = y_actual
        self.model = model
        self.X = X
        
        self.y_predict = self.model.predict(X)
        self.y_predict_proba = self.model.predict_proba(X)[:, 1]
        
        self.class_labels = class_labels
        self.conf_matrix = confusion_matrix(self.y_actual, self.y_predict)
        self.tn, self.fp, self.fn, self.tp = self.conf_matrix.ravel()

    def print_all_matrics(self):
        self._print_accuracy()
        self._print_recall()
        self._print_classification_report(self.y_actual, self.y_predict, self.class_labels)

        plt.rcParams["figure.figsize"] = (11,5)
        
        plt.subplot(1,2,1)
        self._plot_confusion_matrix(self.conf_matrix, self.class_labels)

        if self.y_predict_proba is not None:
            plt.subplot(1,2,2)
            self._plot_roc_curve()
        
        plt.tight_layout()  # automatically adjust params of subplot, https://matplotlib.org/users/tight_layout_guide.html

    def _plot_confusion_matrix(self, cm, classes,
                               normalize=False,
                               title='Confusion matrix',
                               cmap=plt.cm.Blues):
        """
        This function prints and plots the confusion matrix.
        Normalization can be applied by setting `normalize=True`.
        """
        print('---------Confusion matrix and ROC Curve---------')
        '''
        if normalize:
            cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
            print("Normalized confusion matrix")
        else:
            print('Confusion matrix, without normalization')

        print(cm)
        '''

        plt.imshow(cm, interpolation='nearest', cmap=cmap)
        plt.title(title)
        plt.colorbar()
        tick_marks = np.arange(len(classes))
        plt.xticks(tick_marks, classes, rotation=45)
        plt.yticks(tick_marks, classes)

        fmt = '.2f' if normalize else 'd'
        thresh = cm.max() / 2.
        for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
            plt.text(j, i, format(cm[i, j], fmt),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")

        plt.tight_layout()
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        print('')

    def _print_classification_report(self, y_actual, y_predict, class_labels=None):
        print('---------Classification report---------')
        print(classification_report(y_actual, y_predict, target_names=class_labels))
        print('')

    def _print_accuracy(self):
        print('---------Accuracy---------')
        accuracy = (self.tp + self.tn) / (self.tn + self.fp + self.fn + self.tp)
        print(accuracy)
        print('')

    def _print_recall(self):
        print('---------Recall---------')
        recall = recall_score(self.y_actual, self.y_predict)
        print(recall)
        print('')

    def _plot_roc_curve(self):
        # print('---------ROC Curve---------')
        fpr, tpr, thresholds = roc_curve(self.y_actual, self.y_predict_proba, pos_label=1)
        plt.plot(fpr, tpr)
        plt.ylabel('TPR')
        plt.xlabel('FPR')
        
        print('')
```

```python
import tensorflow as tf
from datetime import datetime
from IPython.display import clear_output, Image, display, HTML


def reset_graph(seed=1):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
    
    
def gen_next_batch(batch_size, X, Y, is_bottom_up_data=True):
    m, p = X.shape
    batch_start_index = 0
    batch_num = 0
    while True:
        batch_end_index = min(batch_start_index + batch_size, m)
        if batch_start_index >= m:
            return
            # raise StopIteration
        else:
            batch_num += 1
            # print(batch_num, batch_start_index, batch_end_index)
            if is_bottom_up_data:
                yield (X[batch_start_index:batch_end_index, :], Y[batch_start_index:batch_end_index, :])
            else:
                yield (X[:, batch_start_index:batch_end_index], Y[:, batch_start_index:batch_end_index])
            batch_start_index += batch_size

            
def get_hidden_layers(A0, layer_dims, activation_functions):
    """
    X = tf.placeholder(dtype=tf.float32, shape=[None, n_L0], name='X')
    W1 = tf.Variable(initial_value=tf.random_normal([n_L1, n_L0], stddev=0.01), name='W1')
    L1 = tf.nn.relu(features=tf.matmul(W1, tf.transpose(X)), name='A1')
    W2 = tf.Variable(initial_value=tf.random_normal([n_L2, n_L1], stddev=0.01), name='W2')
    y_proba = tf.nn.sigmoid(x=tf.matmul(W2, L1, name='logits'), name='y_proba')


    # above graph building code is converted as below using 'get_hidden_layers'
    
    A0 = tf.transpose(X)
    layer_dims = [data_X.shape[1], 15, 1]
    activation_functions = [None, tf.nn.relu, tf.nn.sigmoid]
    W, L = get_hidden_layers(A0, layer_dims, activation_functions)
    y_proba = L['L' + str(len(layer_dims) - 1)]
    """
    assert len(layer_dims) == len(activation_functions), 'Hidden layer length miss match'
    
    n_L = len(layer_dims) - 1  # input layer is not counted.

    weight_matrices = {'W0':None}
    layer_output_matrices = {'A0':A0}
    
    for i in range(n_L):
        with tf.name_scope('L' + str(i+1)) as scope:
            weight_matrices['W' + str(i+1)] = tf.Variable(tf.random_normal([layer_dims[i+1], layer_dims[i]], stddev=0.01), name='W' + str(i+1))
            Z_i = tf.matmul(weight_matrices['W' + str(i+1)], layer_output_matrices['A' + str(i)])
            layer_output_matrices['A' + str(i+1)] = activation_functions[i+1](Z_i, name='A' + str(i+1))
    
    return weight_matrices, layer_output_matrices


def strip_consts(graph_def, max_const_size=32):
    """Strip large constant values from graph_def."""
    strip_def = tf.GraphDef()
    for n0 in graph_def.node:
        n = strip_def.node.add()
        n.MergeFrom(n0)
        if n.op == 'Const':
            tensor = n.attr['value'].tensor
            size = len(tensor.tensor_content)
            if size > max_const_size:
                tensor.tensor_content = b"<stripped %d bytes>"%size
    return strip_def


def show_graph(graph_def, max_const_size=32):
    """Visualize TensorFlow graph."""
    if hasattr(graph_def, 'as_graph_def'):
        graph_def = graph_def.as_graph_def()
    strip_def = strip_consts(graph_def, max_const_size=max_const_size)
    code = """
        <script>
          function load() {{
            document.getElementById("{id}").pbtxt = {data};
          }}
        </script>
        <link rel="import" href="https://tensorboard.appspot.com/tf-graph-basic.build.html" onload=load()>
        <div style="height:600px">
          <tf-graph-basic id="{id}"></tf-graph-basic>
        </div>
    """.format(data=repr(str(strip_def)), id='graph'+str(np.random.rand()))

    iframe = """
        <iframe seamless style="width:1200px;height:620px;border:0" srcdoc="{}"></iframe>
    """.format(code.replace('"', '&quot;'))
    display(HTML(iframe))
    

class NNBCM:
    """
    Neural Network Binary Classification Model
    to be used for PrefMatrics class
    
    Binary target variable's name should be 'y_proba'
    """
    def __init__(self, save_path):
        self.save_path = save_path
        
    def predict(self, X_data):
        with tf.Session() as tf_sess:
            tf.train.Saver().restore(tf_sess, self.save_path)
            y_predict = tf_sess.run(y_pred, feed_dict={X:X_data})
            return y_predict.reshape(-1)  # reshape to rank-1 array
    
    def predict_proba(self, X_data):
        with tf.Session() as tf_sess:
            tf.train.Saver().restore(tf_sess, self.save_path)
            y0_proba, y1_proba = tf_sess.run([1-y_proba, y_proba], feed_dict={X:X_data})
            return np.vstack((y0_proba, y1_proba)).T  # n by 2 matrix with 0-proba and 1-proba
    
```

```python
reset_graph()

# logging - part1
now = datetime.now().strftime("%Y%m%d%H%M%S")
root_logdir = "tf_logs"
logdir = "{}/run-{}/".format(root_logdir, now)


# data
data_X = X_train_sm
data_y = y_train_sm.reshape(-1, 1)
n_l0 = data_X.shape[1]

# building graph
with tf.name_scope('y') as scope:
    y = tf.placeholder(dtype=tf.float32, shape=[None, 1], name='y')
    
with tf.name_scope('L0') as scope:
    X = tf.placeholder(dtype=tf.float32, shape=[None, n_l0], name='X')
    A0 = tf.transpose(X)
''

layer_dims = [n_l0, n_l0 * 2, n_l0 * 2, n_l0, 1]
activation_functions = [None, tf.nn.relu, tf.nn.relu, tf.nn.relu, tf.nn.sigmoid]
# activation_functions = [None, tf.nn.tanh, tf.nn.tanh, tf.nn.sigmoid]

W, L = get_hidden_layers(A0, layer_dims, activation_functions)
y_proba = L['A' + str(len(layer_dims) - 1)]

# optimization setting
n_epochs = 1000
batch_size = 5000
n_batches = int(np.ceil(data_X.shape[0] / batch_size))

learning_rate = 0.001
epsilon = 1e-7  # to avoid an overflow when computing the log
# cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(model), axis=1))
cost = -tf.reduce_mean(y * tf.log(y_proba + epsilon) + (1 - y) * tf.log(1 - y_proba + epsilon))
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)


# logging - part2
loss_summary = tf.summary.scalar('log-loss', cost)
file_writer = tf.summary.FileWriter(logdir, tf.get_default_graph())


# model saver
saver = tf.train.Saver()
save_path = '{}/run-{}/model-{}.ckpt'.format(root_logdir, now, now)
```

```python
# training
init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)

    for epoch in range(n_epochs):
        total_cost = 0
        for batch_index, (batch_xs, batch_ys) in enumerate(gen_next_batch(batch_size, data_X, data_y)):
            _, cost_val = sess.run([optimizer, cost], feed_dict={X: batch_xs, y: batch_ys})
            total_cost += cost_val
            
            if batch_index % 5 == 0:
                summary_str = loss_summary.eval(feed_dict={X: batch_xs, y: batch_ys})
                step = (epoch * n_batches + batch_index) * batch_size
                file_writer.add_summary(summary_str, step)

        print('Epoch:', '%04d' % (epoch + 1),
              'Avg. cost =', '{:.3f}'.format(total_cost / n_batches))
    
    _save_path = saver.save(sess, save_path)
''

file_writer.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# initialize random number generator
np.random.seed(123)

# True parameter values
sigma = 5
beta = [3, 5, 9]

# Size of dataset
size = 200
# Predictor variable
X1 = np.linspace(0, 1, size)
X2 = np.linspace(0, 3, size)

# Simulate outcome variable
Y = beta[0] + X1*beta[1] + X2*beta[2] + np.random.randn(size)*sigma

# pylab inline
fig, axes = plt.subplots(1, 2, sharex=True, figsize=(10, 4))
axes[0].scatter(X1, Y)
axes[1].scatter(X2, Y)
axes[0].set_ylabel('Y')
axes[0].set_xlabel('X1')
axes[1].set_xlabel('X2')
```

```python
# Model Specification
from pymc3 import Model, Normal, Gamma
from pymc3 import NUTS, Metropolis, Slice
from pymc3 import sample
from pymc3 import find_MAP
from scipy import optimize

basic_model = Model()

with basic_model:
    # Priors for unknown model parameters
    beta = Normal('beta', mu=0, sd=100, shape=3)
    sigma = Gamma('sigma', alpha=100, beta=100)
    
    # Expected value of outcome
    mu = beta[0] + beta[1]*X1 + beta[2]*X2
    
    # Likelihood(sampling distribution) of observation
    Y_obs = Normal('Y_obs', mu=mu, sd=sigma, observed=Y)

    # obtain starting values via MAP
    #start = find_MAP(fmin=optimize.fmin_powell)
    start = find_MAP()
    #step = Metropolis()
    step = NUTS()
    #step = Slice()
    
    # draw 2000 posterior samples
    trace = sample(3000, step=step, start=start, random_seed=123, chain=2)
    

from pymc3 import traceplot
traceplot(trace[:])

map_estimate = find_MAP(model=basic_model) #Using BFGS optimization algorithm
print(map_estimate)
```

```python
import pymc3 as pm
import numpy as np
import theano
data = [3, 4, 5, 2, 3]

niter = 100
test_model = pm.Model()
with test_model:
    X = theano.shared(np.array(list(range(5))))
    mu = pm.Gamma('mu', 0.1, 0.001)
    Y = pm.Poisson('Y', mu, observed=data)
    
    start = pm.find_MAP()
    trace = pm.sample(niter, start=start, random_seed=2015, progressbar=True)
    pm.traceplot(trace)

trace
```

```python
import math
from scipy.stats import gaussian_kde #Gaussian Kernel Density Estimation

param_name = 'mu'

data = trace.get_values(param_name)
density = gaussian_kde(data)
density.covariance_factor = lambda : .25
density._compute_covariance()

xs = np.linspace(math.floor(min(trace.get_values('mu'))), math.ceil(max(trace.get_values('mu'))), niter)
plt.plot(xs,density(xs), label=param_name)
```
